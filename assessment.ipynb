```python
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.stats as stats 


import numpy as np
import statsmodels.api as sm
import seaborn as sns
```

```python
df = pd.read_csv("assessment_da25.csv")
df.info()
```

```python
df
```

```python


#  Data into groups A and B
group_a = df[df['Variant'] == 'A']
group_b = df[df['Variant'] == 'B']

# Primary metric (Number of add to cart)
primary_metric = 'Number of add to cart'

#  T-test 
t_statistic, p_value = stats.ttest_ind(group_a[primary_metric], group_b[primary_metric])

# Print 
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Comparing p-value to significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the groups based on", primary_metric)
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups based on", primary_metric)

# Visual Inspection for Normality
plt.hist(group_a['Number of add to cart'], bins=20, alpha=0.5, label='Group A')
plt.hist(group_b['Number of add to cart'], bins=20, alpha=0.5, label='Group B')
plt.xlabel('Number of add to cart')
plt.ylabel('Frequency')
plt.title('Histogram of Number of add to cart for each group')
plt.legend()
plt.show()

# Separating data into groups A and B
group_a = df[df['Variant'] == 'A']
group_b = df[df['Variant'] == 'B']

# Calculating mean for each group
mean_a = group_a['Number of add to cart'].mean()
mean_b = group_b['Number of add to cart'].mean()

# A bar plot
plt.bar(['Group A', 'Group B'], [mean_a, mean_b], color=['skyblue', 'lightgreen'])
plt.ylabel('Mean Number of add to cart')
plt.title('Mean Number of add to cart for each group')
plt.show()


# Calculating skewness coefficient
skewness = stats.skew(df['Number of add to cart'])
print("Skewness coefficient:", skewness)

# Plot histogram
plt.hist(df['Number of add to cart'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Number of add to cart')
plt.ylabel('Frequency')
plt.title('Histogram of Number of add to cart')
plt.show()


```

```python
#  CSV data
df = pd.read_csv('assessment_da25.csv')

# Separating data into groups A and B
group_a = df[df['Variant'] == 'A']
group_b = df[df['Variant'] == 'B']

# Secondary metric (Clicks on media)
secondary_metric = 'Clicks on media'

#  T-test
t_statistic, p_value = stats.ttest_ind(group_a[secondary_metric], group_b[secondary_metric])

# Print 
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Comparing p-value to significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the groups based on", secondary_metric)
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups based on", secondary_metric)

# Visual Inspection for Normality
plt.hist(group_a[secondary_metric], bins=20, alpha=0.5, label='Group A')
plt.hist(group_b[secondary_metric], bins=20, alpha=0.5, label='Group B')
plt.xlabel('Clicks on media')
plt.ylabel('Frequency')
plt.title('Histogram of Clicks on media for each group')
plt.legend()
plt.show()

# Calculating mean for each group
mean_a = group_a[secondary_metric].mean()
mean_b = group_b[secondary_metric].mean()

# A bar plot
plt.bar(['Group A', 'Group B'], [mean_a, mean_b], color=['skyblue', 'lightgreen'])
plt.ylabel('Mean Clicks on media')
plt.title('Mean Clicks on media for each group')
plt.show()

# Calculating skewness coefficient
skewness = stats.skew(df['Clicks on media'])
print("Skewness coefficient:", skewness)

# Plot histogram
plt.hist(df['Clicks on media'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Clicks on media')
plt.ylabel('Frequency')
plt.title('Histogram of Clicks on media')
plt.show()
```
